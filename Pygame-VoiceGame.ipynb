```python
import random
import speech_recognition as sr
import pyttsx3

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Set the engine properties
# Set the properties for the male voice
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[-1].id)  # Choose the first male voice in the list
engine.setProperty("rate", 150)   # Set the speaking rate (words per minute)
engine.setProperty("volume", 1) # Set the volume (float from 0.0 to 1.0)

def recognize_speech():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        text = 'Say something!'
        # Speak the text
        engine.say(text)
        engine.runAndWait()
        print(text)
        audio = r.listen(source)
    try:
        return r.recognize_google(audio)
    except sr.UnknownValueError:
        return ""
    except sr.RequestError as e:
        text = 'Could not request results from Google Speech Recognition service; {0}'.format(e)
        # Speak the text
        engine.say(text)
        engine.runAndWait()
        print(text)
        return ""

def play_game():
    text = 'Welcome to the Guess the Number game!'
    # Speak the text
    engine.say(text)
    engine.runAndWait()
    print(text)
    secret_number = random.randint(1, 100)
    text = "I'm thinking of a number between 1 and 100. Can you guess it?"
    # Speak the text
    engine.say(text)
    engine.runAndWait()
    print(text)

    while True:
        guess = recognize_speech()
        list_of_word = ["break", "quit", "exit", "finish", "final", "okay", "ok"]
        if guess == "":
            text = "Sorry, I didn't catch that. Please try again."
            # Speak the text
            engine.say(text)
            engine.runAndWait()
            print(text)
            continue
        elif guess in (list_of_word):
            text = "Exiting the game"
            # Speak the text
            engine.say(text)
            engine.runAndWait()
            print(text)
            break
        try:
            guess = int(guess)
        except ValueError:
            text = "Sorry, I didn't understand that. Please say a number."
            # Speak the text
            engine.say(text)
            engine.runAndWait()
            print(text)
            continue

        if guess < secret_number:
            text = 'Too low. Try again!'
            # Speak the text
            engine.say(text)
            engine.runAndWait()
            print(text)
        elif guess > secret_number:
            text = 'Too high. Try again!'
            # Speak the text
            engine.say(text)
            engine.runAndWait()
            print(text)
        else:
            text = 'Congratulations! You guessed the number!'
            # Speak the text
            engine.say(text)
            engine.runAndWait()
            print(text)
            break

play_game()
```
